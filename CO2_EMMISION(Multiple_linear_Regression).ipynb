```python
import pandas as pd
import matplotlib.pyplot as plt
import  pylab as pl
import numpy as np
from sklearn.linear_model import LinearRegression
%matplotlib inline
```

```python
d=pd.read_csv("/content/CO2 Emissions_Canada.csv")
d
```

```python
d.shape
```

```python
d.info()
```

```python
d.isnull()
```

```python
d.describe()
```

```python
df=d[["Engine Size(L)","Cylinders","Fuel Consumption Comb (mpg)","CO2 Emissions(g/km)"]]
df.head(29)
```

```python
msk=np.random.rand(len(df))<0.8
train=df[msk]
test=df[~msk]
```

```python
lm=LinearRegression()
```

```python
x=np.asanyarray(train[["Engine Size(L)","Cylinders","Fuel Consumption Comb (mpg)"]])
```

```python
y=np.asanyarray(train["CO2 Emissions(g/km)"])
```

```python
lm.fit(x,y)
```

```python
lm.coef_
```

```python
lm.intercept_
```

# ***`PREDICTION`***

```python
y_hat=lm.predict(test[["Engine Size(L)","Cylinders","Fuel Consumption Comb (mpg)"]])
x=np.asanyarray(test[["Engine Size(L)","Cylinders","Fuel Consumption Comb (mpg)"]])
y=np.asanyarray(test[["CO2 Emissions(g/km)"]])
```

```python
y_hat
```

```python
x
```

```python
y
```

```python
print("Mean Square Error:%.2f"%np.mean((y_hat-y)**2))
```

```python
print("Variance score:%.2f"%lm.score(x,y))
```

```python
lm=LinearRegression()
x=np.asanyarray(train[["Engine Size(L)","Cylinders","Fuel Consumption Comb (mpg)"]])
y=np.asanyarray(train[["CO2 Emissions(g/km)"]])
lm.fit(x,y)
```

```python
lm.coef_
```

```python
y_=lm.predict(test[["Engine Size(L)","Cylinders","Fuel Consumption Comb (mpg)"]])
x=np.asanyarray(test[["Engine Size(L)","Cylinders","Fuel Consumption Comb (mpg)"]])
y=np.asanyarray(test[["CO2 Emissions(g/km)"]])
```

```python
print("Residual sum of square(MSE):%.2f"%np.mean(y_-y)**2)
```

```python
print("Variance score:%.2f"%lm.score(x,y))
```
